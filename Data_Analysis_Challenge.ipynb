### Data cleaning 🧹

```python
#Creating a copy
clean_df= df.copy()
```

```python
#standrize  the Course_Type column(lower letter)
clean_df["Course_Type"]= clean_df["Course_Type"].str.lower()
```

```python
#check
clean_df.head()
```

```python
# check if there is any duplicates rows
clean_df.duplicated().sum()
```

```python
# drop duplicate rows
clean_df.drop_duplicates(inplace = True)
```

```python
#check again to make sure duplicates are gone
clean_df.duplicated().sum()
```

```python
#Checking missing values
clean_df.isnull().sum()
```

```python
# Convert 'Rating' to numeric, forcing non-numeric values to NaN
clean_df['Rating'] = pd.to_numeric(clean_df['Rating'], errors='coerce')
```

```python
# Check the info again to see if it changed to float64
print(clean_df.info())
```

```python
#Fill missing values in 'Rating' with the mean to preserve data size
mean_rating= clean_df["Rating"].mean()
clean_df["Rating"] = clean_df["Rating"].fillna(mean_rating)
```

```python
# Verify no more missing values
print(clean_df['Rating'].isnull().sum())
```

```python
# 4) check
clean_df.head(10)
```

```python

# 1) Clean mixed date formats using format='mixed'
clean_df["Enrollment_Date"] = pd.to_datetime(
    clean_df["Enrollment_Date"],
    format="mixed",
    errors="coerce"
)

#2) Remove invalid dates such as 1900-01-01
clean_df = clean_df[clean_df["Enrollment_Date"] != pd.Timestamp("1900-01-01")]

# 3) Reset the index after deletion
clean_df = clean_df.reset_index(drop=True)

# 4) check
clean_df.head(10)
```
